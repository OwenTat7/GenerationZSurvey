# DataSet Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('GenZSurveyXX.csv')
data.head()
```

```python

data['Age'] = data['What is your age?']
data['Education'] = data['What is your current level of education?']
data['Employment Status'] = data['What is your current employment status?']
data['Phone Usage'] = data['How many hours a day do you spend on your phone?']

data.head()
```

```python
data.info()
```

```python
df = data[['Id', 'Age', 'Education', 'Employment Status', 'Phone Usage']]
```

# EDA Demographics (Employment, Age, Education, Phone Usage Hist/Bar)

```python
#Split and normalize
df2 = df['Employment Status'].str.split(';')

#Clean individual items
df2 = df2.apply(lambda statuses: [s.strip().replace("Internship", "Intern") for s in statuses])
#Clean individual items and remove empty strings
df2 = df2.apply(lambda statuses: [s.strip().replace("Internship", "Intern") for s in statuses if s.strip()])


#One-hot encode using MultiLabelBinarizer
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
employment_dummies = pd.DataFrame(mlb.fit_transform(df2), columns=mlb.classes_, index=df.index)

#Combine with original DataFrame
df_combined = pd.concat([df, employment_dummies], axis=1)


```

```python
import matplotlib.pyplot as plt

# Count the number of people in each employment status
status_counts = employment_dummies.sum().sort_values(ascending=False)

# Plot as a bar chart
bars = plt.bar(status_counts.index, status_counts.values, color='skyblue', edgecolor='black')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # x position
        height,                             # y position
        str(int(height)),                   # label text
        ha='center', va='bottom', fontsize=9
    )

# Label the plot
plt.title('Employment Status Distribution')

plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Histogram with labels
n, bins, patches = plt.hist(df['Age'], bins=4, color='lightgreen', edgecolor='black')

# Add title and labels
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Add value labels
for i in range(len(patches)):
    height = n[i]
    plt.text(patches[i].get_x() + patches[i].get_width()/2, height ,
             str(int(height)), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define custom order for education levels
custom_order = [
    "High school",
    "Some college",
    "Associate degree",
    "Bachelor’s degree",
    "Graduate degree"
]

# Get value counts and reindex
edu_counts = df['Education'].value_counts().reindex(custom_order)

# Create bar chart
plt.bar(edu_counts.index, edu_counts.values, color='lightcoral', edgecolor='black')

# Add value labels
for i, v in enumerate(edu_counts.values):
    if pd.notna(v):  # avoid placing text on missing bars
        plt.text(i, v + .2, str(int(v)), ha='center')

# Add labels and title
plt.title('Education Level (Sorted)')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

```

```python
bins1 = ['Less than 2 hours', '2-4 hours', '4-6 hours', 'More than 6 hours']
plt.hist(df['Phone Usage'], bins=4, color='lightgreen', edgecolor='black')

# Add labels and title
plt.title('Phone Usage')
plt.ylabel('Frequency')

plt.xticks(rotation=45)
# Show plot
plt.show()
```

```python
import seaborn as sns

# Ensure column names match your dataset
sns.countplot(data, x='Have you ever encountered a scam online?', 
              hue='How confident are you in identifying a scam?', palette='Set2')

plt.title('Scam Experience vs Scam Confidence')
plt.xlabel('Have You Ever Encountered a Scam?')
plt.ylabel('Number of Respondents')
plt.legend(title='Confidence Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Prepare data: count of comfort levels by age group
comfort_age = pd.crosstab(data['What is your age?'], data['How comfortable are you with sharing personal information online?'])

# Plot stacked bar chart
comfort_age.plot(kind='bar', stacked=True, figsize=(10,6), colormap='tab20')

plt.title('Comfort with Sharing Personal Info by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Respondents')
plt.legend(title='Comfort Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(
    data=data,
    x='What is your current level of education?',
    hue='How confident are you in identifying a scam?',
    palette='Set2'
)

plt.title('Confidence in Identifying Scams by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.legend(title='Confidence Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
